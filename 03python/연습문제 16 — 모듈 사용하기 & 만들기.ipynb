```python
# 1
from datetime import datetime
fixed = datetime(2026, 1, 5, 9, 30, 0)
print(fixed)
```

```python
# 2
print(fixed.year)
print(fixed.month)
print(fixed.day)
```

```python
# 3
print(fixed.hour)
print(fixed.minute)
print(fixed.second)
```

```python
# 4
print(fixed.strftime('%Y.%m.%d'))
```

```python
# 5
print(fixed.strftime('%Y.%m.%d %H:%M:%S'))
```

```python
# 6
print(datetime.now())
```

```python
# 7
now = datetime.now()
print(now.strftime('%Y-%m-%d'))
```

```python
# 8
print(f"{fixed.year}/{fixed.month}/{fixed.day}")
```

```python
# 9
from datetime import datetime

def is_am(dt):
    return dt.hour < 12

print(is_am(datetime(2026,1,5,9,0,0)))
print(is_am(datetime(2026,1,5,15,0,0)))
```
