# 프로그래밍 방법 3가지

# 1. 절차적 프로그래밍
* 프로그래밍을 순차적으로 하는 것
* 코드가 위에서 아래로 절차적으로 실행

```python
nums = [1,2,3,4]
squares = []
for i in nums:
    squares.append(i**2)
print(squares)
```

# 2. 함수형 프로그래밍
* 여러 개의 함수를 작성해서 함수에 기반해 프로그램이 실행
* 코드의 재사용 가능
* 유지보수가 쉬워짐
* 버그 발생률이 낮고 예측 가능성이 높음
* 병렬 처리, 동시성 처리에 강함 - 멀티 코어를 이용할 때 유리

```python
# 리스트의 제곱합 구하기
nums = [1,2,3,4]
squares = list(map(lambda x: x**2, nums)) # lambda라는 익명의 함수 생성
# map (함수, 반복가능한_객체), map도 함수
print(squares)
```

# 3. 객체지향 프로그래밍
* JAVA, C++
* 클래스 기반으로 프로그래밍
* 캡슐화, 상속, 다형성 같은 개념 사용
* 코드를 재사용하는데 특화
* 여러사람이 이용할 때 같은 코드를 독립적으로 사용 가능

```python
class NumberList:
    def __init__(self, numbers):
        self.numbers = numbers
    def squared(self):
        return [x**2 for x in self.numbers]
squared = NumberList([1,2,3,4]).squared()
```

````python
# 함수 내부의 매개변수는 함수 내부에서만 사용됨
result_cal = 0
def add(num, result_cal):
    result_cal += num
    return result_cal
````

````python
# 함수 내부의 매개변수는 함수 내부에서만 사용됨
result_cal = 0
def add(num, result_cal):
    result_cal += num
    print("함수 내부의 result_cal id: ", id(result_cal))
    return result_cal
````

````python
add(3, 0)
print("함수 외부의 result_cal id: ", id(result_cal))
````

* 전역변수를 함수 내부에서 사용할 수 있도록 하는 방법
    * global 전역변수명

````python
result_cal = 0

def add(num):
    global result_cal # result_cal 변수는 위에서 선언된 전역변수 사용할게
    result_cal += num
    print("함수 내부의 result_cal id: ", id(result_cal))
    return result_cal
````

````python
print(add(3))
print("함수 외부의 result_cal id: ", id(result_cal))
````

* 이전 값을 기억하는 계산기를 둘리와 길동이가 같이 쓰려고 할 때

````python
# 둘리의 계산기
result_cal = 0
def add(num):
    global result_cal # result_cal 변수는 위에서 선언된 전역변수 사용할게
    result_cal += num
    print("함수 내부의 result_cal id: ", id(result_cal))
    return result_cal
````

````python
add(3)
````

````python
# 길동이의 계산기
result_cal = 0
def gil_add(num):
    global result_cal # result_cal 변수는 위에서 선언된 전역변수 사용할게
    result_cal += num
    print("함수 내부의 result_cal id: ", id(result_cal))
    return result_cal
````

# 클래스 만들기
* 클래스의 이름은 대문자로 시작하는 카멜 표기법으로 만들어 준다.
* 클래스 안에 정의하는 함수는 메서드라고 부른다.
* 메서드는 인스턴스명.메서드명 형태로 사용
* 인스턴스는 클래스로 만든 제품을 사용자에게 준 것이라고 생각하면 된다.
* 클래스는 공장, 인스턴스는 공장에서 만든 제품을 받은 사용자
* 클래스는 공장에서 만든 계산기, 인스턴스는 만들어진 계산기를 배송받는 사용자

````python
# 공장에서 계산기 설계도를 만듦 => class
class Calc():
    def __init__(self):
        self.result = 0
    def add(self, num):
        self.result += num
        return self.result
````

````python
# 길동이(인스턴스)에게 계산기(class)를 배송
gildong = Calc()
````

````python
# 기능을 사용할 때는 사용자명(길동이, 인스턴스).메서드명(add)
gildong.add(3)
````

````python
# 둘리(인스턴스)에게 계산기(class)를 배송
dul=Calc()
````

````python
dul.add(10)
````

* FourCal에 사칙연산 기능 넣기 

````python
class FourCal():
    # 데이터를 입력받는 역할의 setdata method
    def setdata(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
        print("self.num1", self.num1)
        print("self.num2", self.num2)
    
    def add(self):
        result = self.num1 + self.num2
        return result
    
    def sub(self):
        result = self.num1 - self.num2
        return result
    
    def mul(self):
        result = self.num1 * self.num2
        return result
    
    def div(self):
        result = self.num1 / self.num2
        return result
````

````python
sam = FourCal()
````

````python
sam.add() # setdata에 아직 값 안 넣어줬기 때문
````

````python
sam.setdata(10,5)
````

````python
sam.add()
````

````python
sam.sub()
````

````python
sam.mul()
````

````python
sam.div()
````

# 생성자
* 클래스를 실행해서 인스턴스를 만들 때 자동실행되고 초기값을 받도록 해주는 메서드
* 생성자(constructor)는 객체(인스턴스)가 생성될 때 자동으로 호출되는 메서드
* 클래스 내에서 메서드명으로 \__init\__를 사용하면 그것이 생성자가 된다.

* 생성자를 이용해 초기값을 받는 FourCal2 class

````python
class FourCal2():
    # 생성자
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
        print("self.num1", self.num1)
        print("self.num2", self.num2)
    
    def add(self):
        result = self.num1 + self.num2
        return result
    
    def sub(self):
        result = self.num1 - self.num2
        return result
    
    def mul(self):
        result = self.num1 * self.num2
        return result
    
    def div(self):
        result = self.num1 / self.num2
        return result
````

````python
sally = FourCal2() #  num1과 num2 인수 할당을 안해줌
````

````python
sally = FourCal2(10, 5)
````

````python
sally.add()
````

````python
sally.sub()
````

# 클래스의 상속
* 어떤 클래스를 만들 때 다른 클래스의 기능을 물려받는 것
* 기존 클래스에 기능을 추가할 때
* A클래스를 상속받아 B클래스를 생성한다고 할 때 
* B클래스는 A클래스의 모든 기능을 사용할 수 있다.

* FourCal2를 상속받아 제곱 기능을 추가한 MoreFourCal 만들기

````python
class MoreFourCal(FourCal2): # FourCal2의 기능을 상속받음
    def pow(self):
        result=self.num1 ** self.num2
        return result
````

````python
dul = MoreFourCal()
````

````python
dul = MoreFourCal(10,5)
````

````python
dul.add()
````

````python
dul.sub()
````

````python
dul.mul()
````

````python
dul.div()
````

````python
dul.pow()
````

# 메서드 오버라이딩(method overriding)
* 부모 클래스에서 상속한 메서드를 자식 클래스에서 동일한 이름으로 기능을 재정의 하는 것
* 같은 이름의 메서드의 기능을 바꾸거나 수정할 때 사용

````python
kim = MoreFourCal(100, 0)
````

````python
kim.add()
````

````python
kim.sub()
````

````python
kim.div()
````
