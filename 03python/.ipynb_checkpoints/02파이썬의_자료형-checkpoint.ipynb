# 파이썬의 자료형(data type)
## 파이썬의 자료형은 동적으로 변화
* int: 정수(소숫점이 없는 숫자)
* float: 실수(소숫점이 있는 숫자)
* str: 문자/문자열
* list: 여러개의 자료가 순서대로 저장된 자료형, 모든 자료형을 넣어 만들 수 있음
* tuple: list와 비슷하나 한 번 만들면 수정이 안됨, 모든 자료형을 넣을 수 있음
* dictionary(dict): key : value 쌍으로 이루어진 자료형, 모든 자료형을 넣을 수 있음 
* set: 집합 자료형, 중복이 없는 자료를 만들 때, 모든 자료형을 넣을 수 있음, 대신 중복 불가, {} 안에 자료를 넣어서 만듦

# 리스트  list
* 리스트는 [정수, 실수, 문자, 리스트, 튜플 , 딕셔너리 ] 모두 넣을 수있다.
* 리스트는 여러 자료가 같이 나열된 형태이므로 순서가 있다.
* 인덱싱, 슬라이싱 가능
* 자료의 추가, 삭제가 가능하다.

```python
L1=[1,2.0,'삼',[1,2,3], (4,5,6), {"key":"value"}]
L1
```

```python
len(L1)
```

```python
print(L1, type(L1))
```

```python
L1[1]
```

list는 순서가 있는 자료형

```python
L1.append("리스트 자료형") # 리스트에 추가하는 함수 append
```

```python
print(L1)
```

```python
del L1[6]
```

```python
print(L1)
```

```python
del L1[6]
```

```python
print(L1)
```

## 리스트는 리스트 내의 자료형을 그대로 유지

```python
for item in L1:
    print(item, type(item))
```

# Tuple 튜플 
* 여러가지 자료를 ()안에 넣어서 만드는 자료형
* 모든 data type을 자료로 받을 수 있다.
* 한 번 생성하면 수정이 되지 않는다.
* 튜플이 리스트 보다 코드 실행시 더 빠르다.

```python
L3=(1,2.0,'삼',[1,2,3], (4,5,6), {"key":"value"})
L3
```

```python
print(L3, type(L3))
```

```python
L3.append("튜플") # 튜플은 append로 추가 안됨
```

```python
del L3[-1] # 튜플은 del로 삭제 안됨
```

```python
L3[3][1]
```

# 형변환 함수
* 형변환: data type 다른 data type으로 바꾸는 것

* 정수로 형변환 int()

```python
a=3
b="3"
print(a+b) #숫자 + 문자 이기 때문에 오류가 남
```

```python
a=3
b="3"
print(a+int(b)) #형변환 함수를 이용해 일시적으로 정수로 변환
print("a: ",type(a), "b: ", type(b))
print("a: ",type(a), "int(b): ",type(int(b)))
```

* 실수로 변환 float()

```python
print(b, type(b))
```

```python
a + float(b) # 3+ 3.0 = 6.0
```

```python
c = 4.3
```

```python
a + c 
```

```python
int(a+c)
```

* 문자열로 변환 str()

```python
print(a, type(a))
```

```python
print(str(a), type(str(a)))
```

```python
b
```

```python
str(a) + b
```

* 리스트 형변환 list()
  * 튜플 => 리스트로 변환

```python
t=(1,2,3,4,5)
print(t, type(t))
```

```python
L1 = list(t)
print(L1, type(L1))
```

```python
L1.append(6)
print(L1)
```

* 튜플로 형변환 tuple()

```python
t2=tuple(L1)
print(t2, type(t2))
```

```python
# 주의/ 실수 형태의 소숫점이 있는 문자는 float으로 변환해야 한다. 
d="3.4"
print(d, type(d))
```

```python
print(d, int(d)) 
```

```python
print(d, float(d))
```

# input() : 키보드 입력을 받는 함수
* input("화면에 표시될 메시지")
* 입력 받는 값은 " "으로 구분해서 여러개를 받을 수 있음
* 입력 받는 값은 모두 str 타입의 문자열

```python
input_name = input("이름을 입력해주세요: ")
print(input_name)
```

input함수를 이용해서 숫자 2개를 입력받아 더한 후 출력하세요.

```python
num1=input("숫자 1을 입력하세요: ")
num2=input("숫자 2를 입력하세요: ")
print("숫자 1+ 숫자 2: ", int(num1) + int(num2))
```

* input().split() 으로 여러 데이터를 list 형태로 받기

숫자 5 개를 입력받아서 합을 구하시오

```python
nums=input("숫자를 5개 공백으로 구분해서 입력하세요: ").split()
print(nums, type(nums))
```

```python
int(nums[0]) + int(nums[1]) + int(nums[2]) + int(nums[3]) + int(nums[4])
```
