# 다중 반복문을 사용한 리스트 표현식
```
# 반복문이 2개일 때
[실행할 코드 반복문1 반복문2]
# 반복문이 3개일 때
[실행할 코드 반복문1 반복문2 반복문 3]
```
* 반복문이 3개 이상 중첩되거나 if else 조건이 복잡하게 붙을 경우 일반 문법을 사용
* 코드 해석이 너무 어려워지기 때문

````python
# 2중 반복문으로 구구단 출력하기
gugu=[]
for i in range(2,10):
    for j in range(1,10):
        gugu.append(f"{i} X {j} = {i*j}")
gugu
````

````python
# 리스트표현식으로 2중 반복문 구구단 만들기
gugu2 =[f"{i} X {j} = {i*j}" for i in range(2,10)  for j in range(1,10)]
gugu2
````

````python
# 3중 반복문으로 시계 만들기
clock = []
for si in range(0,24):
    for bun in range(0,60):
        for cho in range(0,60):
            clock.append(f"{si:02d}:{bun:02d}:{cho:02d}")
clock[:10]
````

````python
# 리스트 표현식으로 3중 for문 시계 만들기
clock2 = [f"{si:02d}:{bun:02d}:{cho:02d}" for si in range(0,24) for bun in range(0,60) for cho in range(0,60)]
clock2[:10]
````

# 반복문 + if문의 리스트 표현식 

* 필터링 
```python
for ... if 조건 (for문이 앞에, if 조건문이 뒤에)

* if else + for문의 경우
```python
참일 때 if 조건 else 거짓일 때 for ...(조건 표현식이 앞, for문은 뒤에, if else 함께)
```

# 1) 조건을 이용해 필터링 하기 (if만 쓸 때)

````python
# 1-10 중 짝수만 모으기
# 일반 for + if문
evens=[]
for num in range(1,11):
    if num % 2 == 0:
        evens.append(num)
evens
````

````python
# 리스트 표현식, if문으로 필터링
evens2=[num for num in range(1,11) if num % 2 == 0]
evens
````

## 2) 조건에 따라서 값을 바꾸기 (if~else)

* 1-10 중 짝수면 짝수 홀수면 홀수를 출력

````python
# for + if
labels = []
for num in range(1,11):
    if num % 2 == 0:
        labels.append("짝수")
    else:
        labels.append("홀수")
labels
````

````python
# 리스트표현식 (조건 표현식 + for문)
labels2 = ["짝수" if num % 2 == 0 else "홀수" for num in range(1,11)]     
labels2
````

# 제네레이터 표현식
* 리스트 표현식과 문법이 동일 
* [  ] 아닌 ()로 만듦
* 1회용
* 만들고 나서 반복문으로 내용을 꺼내면 사라짐
* 오직 순차 접근만 가능, 인덱싱, 슬라이싱 불가
* 데이터가 너무 크거나 한번만 쓸 데이터인 경우 사용

# 리스트와 제네레이터 크기 비교

````python
!pip install pympler
````

````python
import sys
from pympler import asizeof
import psutil, os

#메모리 사용량 확인 함수
def show_memory(msg):
    process = psutil.Process(os.getpid())
    print(f"{msg}: {process.memory_info().rss / 1024**2:.2f} MB")

#비교 대상 생성
list_var = [x for x in range(1, 1_000_000)]   # 리스트 내포
gen_var  = (x for x in range(1, 1_000_000))   # 제너레이터 표현식

print("==== sys.getsizeof() 결과 ====")
print("리스트:", sys.getsizeof(list_var))
print("제너레이터:", sys.getsizeof(gen_var))

print("\n==== asizeof.asizeof() 결과 (내부 요소 포함) ====")
print("리스트:", asizeof.asizeof(list_var))
print("제너레이터:", asizeof.asizeof(gen_var))

print("\n==== 프로세스 전체 메모리 사용량 비교 ====")
show_memory("리스트 생성 후")
show_memory("제너레이터 생성 후")
````

````python
ForNum=[]
NatNum=[]
addrCd=[]
gungu=[]
resNm=[]
rnum=[]
sido=[]
yyyymm=[]

for item in data:
    ForNum.append(item['ForNum'])
    NatNum.append(item['NatNum'])
    addrCd.append(item['addrCd'])
    gungu.append(item['gungu'])
    resNm.append(item['resNm'])
    rnum.append(item['rnum'])
    sido.append(item['sido'])
    yyyymm.append(item['yyyymm'])

print('ForNum', ForNum)
print('NatNum', NatNum)
print('addrCd', addrCd)
print('gungu', gungu)
print('resNm', resNm)
print('rnum', rnum)
print('sido', sido)
print('yyyymm', yyyymm)

# 1개의 키에 67개의 데이터가 있는 리스트를 묶어서 dict로 만들기

result=dict(ForNum=ForNum, NatNum=NatNum, gungu=gungu,resNm=resNm, rnum=rnum, sido=sido, yyyymm=yyyymm)
result
````
