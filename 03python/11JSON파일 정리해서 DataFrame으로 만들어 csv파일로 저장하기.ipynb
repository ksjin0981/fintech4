# 강사님 답안

```python
result={}
for test1 in bank_data:
    for key, value in test1.items():
        if key == "별점":
            value=int(value)
        elif key == "리뷰":
            value=value.replace("?", " ").replace(".","").replace("ㅋ","")
        elif key == "개발자 응답":
            if value != None:
                value=value.replace("?", " ").replace(".","").replace("ㅋ","")
        
        result.setdefault(key, []).append(value)
df=pd.DataFrame(result)
df
```

```python
df.info()
```

```python
df.to_csv("./data/kbbank사용자리뷰.csv", index=False, encoding="utf-8-sig")
```

# 전국 치킨집 리스트 데이터 정리하기

```python
with open("./data/전국치킨집리스트.json") as file:
    raw_data = file.read()
    data = json.loads(raw_data)
```

```python
print(data[0])
```

```python
len(data)
```

```python
len(data[0]['header'])
```

```python
len(data[0]['body']['items'])
```

```python
result2={}
for page in data:
    for shop in page['body']['items']:
        for key, value in data[0]['body']['items'][0].items():
            result2.setdefault(key,[]).append(value)
    
df2=pd.DataFrame(result2)
df2
```

```python
print(len(data[0]['header']['columns']))
data[0]['header']['columns']
data[0]['body']['items'][0].keys()
```

```python
result3={}
for page in data[:]:
    for shop in page['body']['items']:
        for key, value in zip(data[0]['header']['columns'], shop.values()):
            result3.setdefault(key,[]).append(value)
    
df3=pd.DataFrame(result3)
df3
```
