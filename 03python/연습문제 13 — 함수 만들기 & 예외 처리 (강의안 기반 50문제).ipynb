```python
# 5
def join_words(a,b):
    joinwords=a+' '+b
    return joinwords

print(join_words('Hello', 'World'))
```

```python
# 6
def power(x, n=2):  # 값을 안 주면 n=2로 쓰이고 값을 주면 그 값이 기본값을 덮어씀
    return x**n

print(power(5))
print(power(2, 3))
```

```python
# 7
def say(name, msg='안녕하세요'):
    print(f"{msg}, {name}!")
say('지현')
say('지현', '반가워요')
```

```python
# 8 # 질문
def sum_and_diff(a, b):
    return (a+b, a-b)

s, d=sum_and_diff(10, 3)
print(s)
print(d)
```

```python
# 9
def quot_rem(a, b):
    return a//5, a%b

q, r = quot_rem(17, 5)
print(q, r)
```

```python
# 10
def print_positive(n):
    if n<0:
        print("음수는 처리하지 않습니다")
        return # return은 “값을 돌려준다” + “함수를 끝낸다”
                # 그래서 값 없이 return만 써도 함수 종료 의미는 같다
    else:
        print(n)
        
print_positive(-1)
print_positive(7)
```

```python
# 13
def profile(name, age, city):
    print(f"이름: {name}\n나이: {age} \n도시: {city}")

profile(city='Seoul', age=25, name='Alice')
```

```python
# 14 
# 반복문 써서 구하면 됨 !
def sum_all(*args):
    hap=0
    for i in args:
        hap+=i
    return hap
print(sum_all(1, 2, 3, 4))
```
