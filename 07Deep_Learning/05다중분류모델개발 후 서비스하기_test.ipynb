# 테스트 데이터에 훈련 데이터에 했던 작업 동일하게 적용

* 종속변수 LabelEncoder로 인코딩

```python
X_test = data.drop('와인_종류', axis=1)
y_test = data['와인_종류']
```

```python
import joblib
```

```python
rs = joblib.load("./model/Robust_info.joblib")
le = joblib.load("./model/labelencoder.joblib")
```
