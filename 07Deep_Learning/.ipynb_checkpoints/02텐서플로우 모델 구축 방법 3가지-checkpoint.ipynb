```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import koreanize_matplotlib
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/haram4th/ablearn/main/Taitanic_train.csv")
data.head()
```

```python
data.columns
```

```python
data = data[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Embarked']]
data.head()
```

```python
data['family']=data['SibSp'] + data['Parch']
data.head()
```

```python
data = data.drop(['SibSp', 'Parch'], axis=1)
data.head()
```

```python
X = data.drop('Survived', axis=1)
y = data['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, stratify=y, random_state=42)
```

```python
X_train.info()
```

# л”Ҙлҹ¬лӢқмӢңм—җлҸ„ лҚ°мқҙн„° мҠӨмјҖмқјмқ„ л§һм¶° мЈјм–ҙм•ј н•Ё
* MinMaxScalerлҘј мЈјлЎң мӮ¬мҡ©, StandardScalerлҸ„ мӮ¬мҡ©, мқҙмғҒк°’мқҙ л§ҺмқҖ кІҪмҡ° RobustScaler мӮ¬мҡ©
* мқҙм§„л¶„лҘҳ: MinMaxScaler
* м—°мҶҚк°’мҳҲмёЎ: StandardScaler

```python
X_train.columns
```

```python
num_cols = ['Age', 'family']
cat_cols = ['Pclass', 'Sex', 'Embarked']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer # кІ°мёЎм№ҳ мұ„мӣҢмЈјлҠ” лҸ„кө¬
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer # м»¬лҹјлі„лЎң лӢӨлҘё м „мІҳлҰ¬лҘј м Ғмҡ©н•ҳлҠ” лҸ„кө¬
```

```python
num_pipe = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='mean')),
    ('mmx', MinMaxScaler())
])

cat_pipe = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
])
# drop='first': мІ«лІҲм§ё м№ҙн…Ңкі лҰ¬ м ңкұ° (лӢӨмӨ‘кіөм„ м„ұ л¬ём ң л°©м§Җ)
# handle_unknown='ignore': н•ҷмҠөн•  л•Ң м—ҶлҚҳ мғҲлЎңмҡҙ м№ҙн…Ңкі лҰ¬к°Җ лӮҳмҳӨл©ҙ м—җлҹ¬ лӮҙм§Җ м•Ҡкі  л¬ҙмӢң
# sprase_output=False: м¶ңл Ҙмқ„ нқ¬мҶҢн–үл ¬мқҙ м•„лӢҲлқј мқјл°ҳ л°°м—ҙлЎң л°ҳнҷҳ
prep = ColumnTransformer(transformers=[
    ('num_cols', num_pipe, num_cols),
    ('cat_cols', cat_pipe, cat_cols)
], remainder='drop')

X_train_ppd = prep.fit_transform(X_train)
X_test_ppd = prep.transform(X_test)
```

# Tensorflow Sequential APIлҘј мӮ¬мҡ©н•ң мҲңм°Ём Ғ лӘЁлҚё мғқм„ұ л°Ҹ л¶„м„қ
* Sequential: мқёкіөмӢ кІҪл§қмқ„ мҲңм°Ём ҒмңјлЎң мһ‘м„ұн•  л•Ң мӮ¬мҡ©
* Dense: мёө, л Ҳмқҙм–ҙ, мһ…л Ҙмёө, мқҖлӢүмёө, м¶ңл Ҙмёө
    * мһ…л Ҙмёө: лӘЁлҚём—җ лҚ°мқҙн„°лҘј л„ЈлҠ” мёө
    * мқҖлӢүмёө: мһ…л Ҙ лҚ°мқҙн„°лҘј кі„мӮ°н•ҳкі  нҢЁн„ҙмқ„ н•ҷмҠөн•ҳлҠ” мёө
    * м¶ңл Ҙмёө: лӘЁлҚёмқҙ мөңмў… кІ°кіјлҘј лӮҙлҠ” мёө

```python
import keras 
import tensorflow as tf
from keras import Input
from keras.models import Sequential
from keras.layers import Dense
```

* лүҙлҹ°: мһ…л Ҙк°’мқ„ л°ӣм•„ кі„мӮ°н•ҙм„ң н•ҳлӮҳмқҳ м¶ңл Ҙк°’мқ„ л§Ңл“ңлҠ” мһ‘мқҖ кі„мӮ° лӢЁмң„ (мһ…л Ҙ -> кі„мӮ° -> м¶ңл Ҙ)
* мһ…л ҘмёөмқҖ л°ҳл“ңмӢң лҸ…лҰҪліҖмҲҳмқҳмҲҳ(input_dim)лҘј м •нҷ•нһҲ м§Җм •
* мӢ кІҪл§қ лӘЁлҚёмқҳ мқҖлӢүмёөмқҖ 2~3к°ң м •лҸ„лЎң н•ңлӢӨ.
* Denseмқҳ л…ёл“ңмҲҳлҠ” 2мқҳ л°°мҲҳмқҳ м§Җм • 2 4 8 16 256
* Denseмқҳ л…ёл“ңлҠ” нҒ° мҲҳм—җм„ң мһ‘мқҖ мҲҳлЎң мӨ„м—¬лӮҳк°„лӢӨ.
* мқҖлӢүмёөмқҳ нҷңм„ұнҷ” н•ЁмҲҳ: Relu
    * нҷңм„ұнҷ” н•ЁмҲҳ: лүҙлҹ°мқҳ м¶ңл Ҙк°’мқ„ ліҖнҷҳн•ҳлҠ” н•ЁмҲҳ (мӢ кІҪл§қмқҙ ліөмһЎн•ң нҢЁн„ҙмқ„ н•ҷмҠөн•ҳкІҢ л§Ңл“Өкё° мң„н•ҙ)
* м¶ңл ҘмёөмқҖ мқҙм§„л¶„лҘҳмқј кІҪмҡ° sigmoid, лӢӨмӨ‘л¶„лҘҳмқј кІҪмҡ° softmax
* нҡҢк·Җл¶„м„қмқј кІҪмҡ°(мў…мҶҚліҖмҲҳк°Җ м—°мҶҚнҳ•(мҲҳм№ҳнҳ•)мқё кІҪмҡ°) м¶ңл Ҙмёөм—җ нҷңм„ұнҷ” н•ЁмҲҳ м—Ҷмқҙ л…ёл“ңлҘј 1лЎң м§Җм •

```python
X_train_ppd.shape[1] # лҸ…лҰҪліҖмҲҳ мҲҳ кі„мӮ° (м—ҙмқҳ к°ңмҲҳ)
```

```python
# # мҳӣлӮ  лІ„м „ мҪ”л“ң
# from tensorflow.keras.models import Sequential
# from tensorflow.keras.layers import Dense

# model = Sequential()

# model.add(Dense(64,
#                 activation='relu',
#                 input_dim=X_train_ppd.shape[1]))  # мһ…л Ҙмёө

# model.add(Dense(32, activation='relu'))  # мқҖлӢүмёө
# model.add(Dense(16, activation='relu'))  # мқҖлӢүмёө

# model.add(Dense(1, activation='sigmoid'))  # м¶ңл Ҙмёө
```

```python
# keras 3.0 кё°мӨҖ мӢ кІҪл§қ лӘЁлҚё м •мқҳ
model = Sequential()
model.add(Input(shape=(X_train_ppd.shape[1],))) # мһ…л Ҙмёө
model.add(Dense(64, activation = 'relu')) # мқҖлӢүмёө
model.add(Dense(32, activation = 'relu')) # мқҖлӢүмёө
model.add(Dense(16, activation = 'relu')) # мқҖлӢүмёө
model.add(Dense(1, activation='sigmoid')) # м¶ңл Ҙмёө, мқҙм§„л¶„лҘҳмқҙлҜҖлЎң sigmoid
```

```python
model.summary()
```

* params: к°Ғ л Ҳмқҙм–ҙк°Җ к°Җм§Җкі  мһҲлҠ” н•ҷмҠө к°ҖлҠҘн•ң нҢҢлқјлҜён„° к°ңмҲҳ
    * params = к°ҖмӨ‘м№ҳ + bias
    
* мІ«лІҲм§ё л Ҳмқҙм–ҙ (None, 64)
    * лүҙлҹ° 64к°ң 
    * Param # 512к°ң 
        * (мһ…л Ҙл…ёл“ң * м¶ңл Ҙл…ёл“ң) + м¶ңл Ҙл…ёл“ң
        * (7 * 64) + 64
* л‘җлІҲм§ё л Ҳмқҙм–ҙ (None, 32)
    * мқҙ л Ҳмқҙм–ҙмқҳ мһ…л ҘмқҖ мқҙм „ л Ҳмқҙм–ҙмқҳ 64к°ң м¶ңл Ҙ
        * (64 * 32) + 32

#  лӘЁлҚёмқҳ нҢҢлқјлҜён„° м„ёнҢ… model.compile()
* loss: мҳӨм°Ё кі„мӮ° м§Җн‘ң (лӘЁлҚёмқҙ м–јл§ҲлӮҳ нӢҖл ёлҠ”м§Җ кі„мӮ°)
  * мқҙм§„л¶„лҘҳ: binary_crossentropy
  * лӢӨмӨ‘л¶„лҘҳ: categorical_crossentropy - yк°Җ one-hot encoding
  * лӢӨмӨ‘л¶„лҘҳ2: sparse_categorical_crossentropy - yк°Җ label encoding
  * м—°мҶҚнҳ•лҚ°мқҙн„°(нҡҢк·Җл¶„м„қ): mse
* optimizer: learning_rate ( к·ё нӢҖлҰ° м •лҸ„лҘј ліҙкі  м–ҙл–»кІҢ кі м№ м§Җ кІ°м •)
  * adamмқ„ к°ҖмһҘ л§Һмқҙ мӮ¬мҡ©
* metrics: нҸүк°Җм§Җн‘ң accuracy

```python
pd.get_dummies(y_train)
```

```python
# лӘЁлҚёмқҳ нҢҢлқјлҜён„° м„ёнҢ…
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

# лӘЁлҚё нӣҲл Ё model.fit(X_train, y_train)
* epochs: м „мІҙ лҚ°мқҙн„°лҘј н•ң лІҲ нӣҲл Ён•ҳлҠ” мЈјкё°, 10-100 мӮ¬мқҙ
    * м „мІҙ лҚ°мқҙн„° н•ҷмҠө нҡҹмҲҳ
* batch_size: м „мІҙ лҚ°мқҙн„°лҘј мӘјк°ңм„ң нӣҲл Ё, лҚ°мқҙн„°мқҳ л¬¶мқҢ, л©”лӘЁлҰ¬м—җ л§һм¶°м„ң нҒ¬кё° мЎ°м Ҳ(2мқҳ л°°мҲҳ) 32, 64лЎң мӘјк°ңм„ң л„ЈлҠ” кІғ м¶”мІң
    * н•ңлІҲм—җ лӘЁлҚём—җ л„ЈлҠ” лҚ°мқҙн„° к°ңмҲҳ
* validation_data: кІҖмҰқ лҚ°мқҙн„°лЎң лӘЁлҚё кІҖмҰқ мӢӨмӢң

```python
history = model.fit(X_train_ppd, y_train, epochs=300, batch_size=16,
                   validation_data=(X_test_ppd, y_test))
```

# л”Ҙлҹ¬лӢқ лӘЁлҚё м„ұлҠҘ нҸүк°Җ
* л”Ҙлҹ¬лӢқм—җм„ң лӘЁлҚёмқҳ м„ұлҠҘмқ„ нҸүк°Җн•ҳлҠ” м§Җн‘ңлҠ” loss
* ліҙнҶө lossк°Җ лӮ®мңјл©ҙ accuracyк°Җ мҳ¬лқјк°Җм§Җл§Ң л№„лЎҖн•ҳм§Җ м•Ҡмқ„ л•ҢлҸ„ мһҲмқҢ.
* accuracyмҷҖ loss к°„мқҳ м°Ёмқҙк°Җ мһҲмқ„ л•Ң lossк°Җ лӮ®мқҖ лӘЁлҚёмқҙ мҡ°мҲҳн•ң лӘЁлҚё

* val lossк°Җ к°ҖмһҘ лӮ®мқҖ м§Җм җмқҙ к°ҖмһҘ мўӢмқҖ лӘЁлҚё
    * м•Ҫ 20~40 epoch мӮ¬мқҙ

# Functional APIлҘј мқҙмҡ©н•ң н•ЁмҲҳнҳ• лӘЁлҚё
* Sequential APIлҠ” лӢЁмҲңнһҲ мёөмқ„ м—¬лҹ¬ к°ң мҢ“лҠ” нҳ•нғңлқј ліөмһЎн•ң лӘЁлҚё мғқм„ұм—җ н•ңкі„к°Җ мһҲмқҢ.
* Functional APIлҠ” мһ…л Ҙмёөкіј м¶ңл Ҙмёөмқ„ мӮ¬мҡ©мһҗк°Җ м§Ғм ‘ м •мқҳ к°ҖлҠҘ
* лӢӨмӨ‘мһ…л Ҙ(Multi-input), лӢӨмӨ‘м¶ңл Ҙ(Multi-output)л“ұ ліөмһЎн•ң лӘЁлҚёмқ„ м •мқҳ н•  мҲҳ мһҲмқҢ.
* Input(shape=(лҸ…лҰҪліҖмҲҳ мҲҳ))лЎң мһ…л Ҙмёө м •мқҳ
* мқҙм „ мёөмқ„ лӢӨмқҢ мёөмқҳ мһ…л ҘмңјлЎң мӮ¬мҡ©
* model()м—җ мһ…л Ҙкіј м¶ңл Ҙмқ„ м •мқҳ

```python
from keras.layers import Input, Dense
from keras.models import Model
```

```python
# # мӢ кІҪл§қ лӘЁлҚё м •мқҳ kearas 3.0, tensorflow 2.19 кё°мӨҖ
# model = Sequential()
# model.add(Input(shape=(X_train_ppd.shape[1],))) # мһ…л Ҙмёө
# model.add(Dense(64, activation='relu')) # мқҖлӢүмёө
# model.add(Dense(32, activation='relu')) # мқҖлӢүмёө
# model.add(Dense(16, activation='relu')) # мқҖлӢүмёө
# model.add(Dense(1, activation='sigmoid')) # м¶ңл Ҙмёө, мқҙм§„л¶„лҘҳмқҙлҜҖлЎң sigmoid
```

```python
# мһ…л Ҙмёө
inputs = Input(shape=(X_test_ppd.shape[1],))

# мқҖлӢүмёө
x = Dense(64, activation='relu')(inputs)
x = Dense(32, activation='relu')(x)
x = Dense(16, activation='relu')(x)

# м¶ңл Ҙмёө
outputs = Dense(1, activation='sigmoid')(x)

# лӘЁлҚё м •мқҳ
model = Model(inputs=inputs, outputs=outputs)
model.summary()
```

```python
from sklearn.metrics import classification_report
```

```python
# лӘЁлҚёмқҳ нҢҢлқјлҜён„° м„ёнҢ…
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(X_train_ppd, y_train, epochs=100, batch_size=16,
                   validation_data=(X_test_ppd, y_test))
plt.figure(figsize=(8,5))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend(['train', 'val'])
plt.show()
pred = model.predict(X_test_ppd)
pred
result = pd.DataFrame(dict(y_test=y_test, pred=pred[:,0]))
result['pred'] =  result['pred'].apply(lambda x: 1 if x >= 0.5 else 0)
print(classification_report(result['y_test'], result['pred']))
```

# classнҳ•мңјлЎң л§Ңл“Өкё° (Model Subclass API)
* class нҳ•нғңмқҳ лӘЁлҚё м •мқҳ л°©лІ•
* kerasмқҳ Modelмқ„ мғҒмҶҚл°ӣм•„ мһ‘м„ұ

```python
# # мһ…л Ҙмёө
# inputs = Input(shape=(X_test_ppd.shape[1],))

# # мқҖлӢүмёө
# x = Dense(64, activation='relu')(inputs)
# x = Dense(32, activation='relu')(x)
# x = Dense(16, activation='relu')(x)

# # м¶ңл Ҙмёө
# outputs = Dense(1, activation='sigmoid')(x)

# # лӘЁлҚё м •мқҳ
# model = Model(inputs=inputs, outputs=outputs)
# model.summary()
```

```python
class Titanic(Model):
    def __init__(self): # л Ҳмқҙм–ҙ мғқм„ұ
        super().__init__() # мһ…л Ҙмёө
        self.dense1 = Dense(128, activation='relu') # мқҖлӢүмёө
        self.dense2 = Dense(64, activation='relu')  # мқҖлӢүмёө
        self.dense3 = Dense(32, activation='relu')  # мқҖлӢүмёө
        self.dense4 = Dense(16, activation='relu')  # мқҖлӢүмёө
        self.outputs = Dense(1, activation='sigmoid') # м¶ңл Ҙмёө
        
    def call(self, inputs): # мӢӨм ң лӘЁлҚё
        x = self.dense1(inputs)
        x = self.dense2(x)
        x = self.dense3(x)
        x = self.dense4(x)
        return self.outputs(x)
```

```python
model = Titanic()
print(model.summary())
```

```python
# лӘЁлҚёмқҳ нҢҢлқјлҜён„° м„ёнҢ…
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(X_train_ppd, y_train, epochs=100, batch_size=16,
                   validation_data=(X_test_ppd, y_test))
plt.figure(figsize=(8,5))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend(['train', 'val'])
plt.show()

pred = model.predict(X_test_ppd)
pred
result = pd.DataFrame(dict(y_test=y_test, pred=pred[:,0]))
result['pred'] =  result['pred'].apply(lambda x: 1 if x >= 0.5 else 0)
print(classification_report(result['y_test'], result['pred']))
```
