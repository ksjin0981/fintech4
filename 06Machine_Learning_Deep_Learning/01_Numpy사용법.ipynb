# np.where(조건, 참, 거짓) 3항 연산

# 배열 합치기 np.concatenate()

```python
a = np.array([[1,2], [3,4]])
b = np.array([[5,6], [7,8]])
print(a, a.ndim, a.shape)
print(b, b.ndim, b.shape)
```

* np.concatenate((a,b), axis = 0/1) 기본값은 axis=0
* axis=0 아래로 합쳐짐 (행이 늘어남)
* axis=1 옆으로 합쳐짐 (열이 늘어남)

```python
# axis=0 아래로 늘어나면서 합쳐짐
np.concatenate((a,b))
```

```python
# axis=1 옆으로 늘어나면서 합쳐짐
np.concatenate((a,b), axis=1)
```

# 축 (axis) 기반 집계 연산 np.sum()
* axis = 0 세로 (열) 단위 연산
* axis = 1 가로(행) 단위 연산

```python
mat = np.array([[1,2,3], [4,5,6]])
print(mat, mat.ndim, mat.shape)
```

```python
print("전체요소의 합: ", np.sum(mat))
print("열별 합 (axis=0)", np.sum(mat, axis=0))
print("행별 합 (axis=1)", np.sum(mat, axis=1))
```

# np.mean() 평균

```python
print("전체요소의 평균: ", np.mean(mat))
print("열별 평균 (axis=0)", np.mean(mat, axis=0))
print("행별 평균 (axis=1)", np.mean(mat, axis=1))
```

# 누적합/누적곱 (cumsum/cumprod)
* cumsum: 앞에서부터 계속 더한 결과 (현금 잔고 추이 등)
* cumprod: 앞에서부터 계속 곱한 결과 (누적 성장률 등)

```python
cashflow = np.array([100, -20, 50, -10])
print("현금흐름: ", cashflow)
print("누적잔고: ", np.cumsum(cashflow))
```

```python
# 매년 매출 성장 비율
rates = np.array([10.5, 13, 0.5, 1.3])
print("연별 성장률: ", rates)
print("누적 성장률: ", np.cumprod(rates))
```

# 분산 (var), 표준편차 (std)

```python
classA = np.array([70, 72, 68, 69, 71])
classB = np.array([50, 90, 30, 100, 10])
```

```python
print("A반 평균: ", np.mean(classA))
print("B반 평균: ", np.mean(classB))
```

```python
print("A반 표준편차: ", np.std(classA))
print("B반 표준편차: ", np.std(classB))
```

```python
print("A반 분산: ", np.var(classA))
print("B반 분산: ", np.var(classB))
```
