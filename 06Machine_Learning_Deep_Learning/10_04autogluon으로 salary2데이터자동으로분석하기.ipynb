````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib

from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_auc_score
)

from autogluon.tabular import TabularPredictor

pd.set_option("display.max_columns", 50)
````

## 3. ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸° ë° ê¸°ë³¸ ì „ì²˜ë¦¬

### 3.1 salary2.csv ë¶ˆëŸ¬ì˜¤ê¸°

- Adult Income ê¸°ë°˜ì˜ ì—°ë´‰ ë°ì´í„°ì…‹ì„ ì‚¬ìš©í•œë‹¤.
- íŒŒì¼ ìœ„ì¹˜ëŠ” `./data/salary2.csv` ë¼ê³  ê°€ì •í•œë‹¤.


````python
# ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
data = pd.read_csv("./data/salary2.csv")

print("ë°ì´í„° í¬ê¸°:", data.shape)
data.head()
````

### 3.2 ì¤‘ë³µ ì œê±° & ê²°ì¸¡ê°’ ì œê±°

- **ì¤‘ë³µ í–‰(drop_duplicates)**: ê°™ì€ ì‚¬ëžŒì´ ì—¬ëŸ¬ ë²ˆ ê¸°ë¡ë˜ì—ˆì„ ìˆ˜ ìžˆì–´ì„œ ì œê±°
- **ê²°ì¸¡ê°’ ì œê±°(dropna)**: ê²°ì¸¡ì¹˜ê°€ ìžˆëŠ” í–‰ì€ ê°„ë‹¨í•˜ê²Œ ì œê±° (ìˆ˜ì—…ìš© ì˜ˆì œë¼ì„œ)
- ì¸ë±ìŠ¤ë¥¼ ë‹¤ì‹œ 0ë¶€í„° ìž¬ì„¤ì •í•œë‹¤.


````python
# ì¤‘ë³µ ì œê±°
data = data.drop_duplicates()

# ê²°ì¸¡ê°’ ì œê±°
data = data.dropna()

# ì¸ë±ìŠ¤ ìž¬ì„¤ì •
data = data.reset_index(drop=True)

print("ì „ì²˜ë¦¬ í›„ ë°ì´í„° í¬ê¸°:", data.shape)
data.head()
````

### 3.3 íƒ€ê¹ƒ ë³€ìˆ˜ `class` ì´ì§„ ë³€í™˜

ì›ë³¸ ë°ì´í„°ì˜ `class` ì»¬ëŸ¼ì€ ë¬¸ìžì—´ í˜•íƒœì´ë‹¤. (ì˜ˆ: `' >50K'`, `' <=50K'`)

- `' >50K'`  â†’  **1 (ì—°ë´‰ 50K ì´ˆê³¼)**  
- ê·¸ ì™¸(ì˜ˆ: `' <=50K'`) â†’ **0 (ì—°ë´‰ 50K ì´í•˜)**

ì´ë ‡ê²Œ ë°”ê¿”ì•¼ ë¨¸ì‹ ëŸ¬ë‹ ëª¨ë¸ì´ **ì´ì§„ ë¶„ë¥˜ ë¬¸ì œ(binary classification)**ë¡œ ë‹¤ë£° ìˆ˜ ìžˆë‹¤.


````python
# class ì»¬ëŸ¼ì„ 0/1ë¡œ ë³€í™˜
data['class'] = data['class'].apply(lambda x: 1 if x.strip() == '>50K' or x.strip() == '>50K.' or x.strip() == '>50K' else 0)

# íƒ€ê¹ƒ ë¶„í¬ í™•ì¸
data['class'].value_counts()
````

### 3.4 ê°„ë‹¨í•œ EDA (íƒ€ê¹ƒ ë¹„ìœ¨ í™•ì¸)

ì—°ë´‰ì´ 50K ì´ˆê³¼ì¸ ì‚¬ëžŒ(1)ê³¼ ê·¸ë ‡ì§€ ì•Šì€ ì‚¬ëžŒ(0)ì˜ ë¹„ìœ¨ì„ í™•ì¸í•˜ì—¬  
**í´ëž˜ìŠ¤ ë¶ˆê· í˜•(class imbalance)**ì´ ìžˆëŠ”ì§€ ì‚´íŽ´ë³¸ë‹¤.


````python
target_counts = data['class'].value_counts().rename(index={0: 'â‰¤50K', 1: '>50K'})
target_counts
````

````python
# ë§‰ëŒ€ ê·¸ëž˜í”„ë¡œ ì‹œê°í™”
plt.figure(figsize=(5,4))
target_counts.plot(kind='bar')
plt.title("ì—°ë´‰ êµ¬ê°„ë³„ ìƒ˜í”Œ ìˆ˜")
plt.ylabel("ìƒ˜í”Œ ìˆ˜")
plt.xticks(rotation=0)
plt.show()
````

## 4. í•™ìŠµìš© / í‰ê°€ìš© ë°ì´í„° ë¶„ë¦¬

ë¨¸ì‹ ëŸ¬ë‹ ëª¨ë¸ì˜ **ì¼ë°˜í™” ì„±ëŠ¥**ì„ í‰ê°€í•˜ê¸° ìœ„í•´ ë°ì´í„°ì…‹ì„ ë‚˜ëˆˆë‹¤.

- `train_df` : ëª¨ë¸ í•™ìŠµì— ì‚¬ìš© (60%)
- `test_df` : ìµœì¢… ì„±ëŠ¥ í‰ê°€ì— ì‚¬ìš© (40%)
- `stratify=y` ì˜µì…˜ìœ¼ë¡œ íƒ€ê¹ƒ ë¹„ìœ¨ì„ ìœ ì§€í•œ ì±„ë¡œ ë‚˜ëˆˆë‹¤.


````python
# ì „ì²´ ë°ì´í„° ë³µì‚¬
df = data.copy()

# í•™ìŠµ/í…ŒìŠ¤íŠ¸ ë¶„í• 
train_df, test_df = train_test_split(
    df,
    test_size=0.4,
    stratify=df['class'],
    random_state=10
)

print("train_df shape:", train_df.shape)
print("test_df shape:", test_df.shape)

train_df.head()
````

## 5. AutoGluon Tabular ì†Œê°œ

`AutoGluon`ì€ **AutoML ë¼ì´ë¸ŒëŸ¬ë¦¬**ë¡œ, ì•„ëž˜ ìž‘ì—…ë“¤ì„ ìžë™ìœ¼ë¡œ í•´ì¤€ë‹¤.

- ë²”ì£¼í˜•/ìˆ˜ì¹˜í˜• ë³€ìˆ˜ ìžë™ ì¸ì½”ë”©
- ì—¬ëŸ¬ ëª¨ë¸(GBM, CatBoost, XGBoost ë“±) ìžë™ íƒìƒ‰
- í•˜ì´í¼íŒŒë¼ë¯¸í„° íŠœë‹
- ì—¬ëŸ¬ ëª¨ë¸ì„ ì•™ìƒë¸”í•˜ì—¬ ìµœì¢… ì„±ëŠ¥ í–¥ìƒ

ìš°ë¦¬ëŠ” ë‹¨ì§€:
1. **ë°ì´í„°í”„ë ˆìž„** ì „ì²´ì™€  
2. **íƒ€ê¹ƒ ì»¬ëŸ¼ ì´ë¦„(label)** ë§Œ ë„˜ê²¨ì£¼ë©´ ëœë‹¤.

ì´ì œ `TabularPredictor`ë¥¼ ì´ìš©í•´ì„œ ëª¨ë¸ì„ í•™ìŠµí•´ë³´ìž.


### 5.1 TabularPredictor ì„¤ì • ë° í•™ìŠµ

- `label='class'` : ì˜ˆì¸¡í•  íƒ€ê¹ƒ ì»¬ëŸ¼ ì´ë¦„  
- `problem_type='binary'` : ì´ì§„ ë¶„ë¥˜ ë¬¸ì œ  
- `eval_metric='f1'` : ë¶ˆê· í˜• ë°ì´í„°ì— ìœ ë¦¬í•œ F1 ì ìˆ˜ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ìµœì  ëª¨ë¸ ì„ íƒ  
- `presets="medium_quality_faster_train"` : ì ë‹¹í•œ ì†ë„ + ê´œì°®ì€ ì„±ëŠ¥  
- `time_limit=600` : ìµœëŒ€ í•™ìŠµ ì‹œê°„(ì´ˆ) ì„¤ì • (í™˜ê²½ì— ë”°ë¼ ì¡°ì • ê°€ëŠ¥)


````python
save_path = "autogluon_salary_model/"  # ëª¨ë¸ ë° ë¡œê·¸ ì €ìž¥ ê²½ë¡œ

predictor = TabularPredictor(
    label='class',           # íƒ€ê¹ƒ ì»¬ëŸ¼ ì´ë¦„
    problem_type='binary',   # ì´ì§„ë¶„ë¥˜
    eval_metric='f1',        # í‰ê°€ ì§€í‘œ (ë¶ˆê· í˜• ë°ì´í„°ì— ìœ ë¦¬)
    path=save_path
)

# presets:
# - "medium_quality_faster_train": ì ë‹¹í•œ ì†ë„ + ê´œì°®ì€ ì„±ëŠ¥
# - "best_quality": ì‹œê°„ ì˜¤ëž˜ ê±¸ë¦¬ì§€ë§Œ ìµœìƒ ì„±ëŠ¥  (ë°ì´í„° í¬ê¸°ì— ë”°ë¼ ì¡°ì •)
# time_limit: ì´ˆ ë‹¨ìœ„ ì œí•œ (ì˜ˆ: 600 = 10ë¶„)
predictor = predictor.fit(
    train_data=train_df,
    presets="best_quality",
    time_limit=120,  # í•„ìš”ì— ë”°ë¼ ì¦ê°€/ê°ì†Œ
    dynamic_stacking=False,
    verbosity=2
)
````

## 6. ë¦¬ë”ë³´ë“œ(LeaderBoard)ë¡œ ëª¨ë¸ ë¹„êµ

`predictor.leaderboard(test_df)`ë¥¼ ì‚¬ìš©í•˜ë©´,  
AutoGluonì´ í•™ìŠµí•œ **ì—¬ëŸ¬ ëª¨ë¸ë“¤ì˜ ì„±ëŠ¥ì„ í•œ ë²ˆì— ë¹„êµ**í•  ìˆ˜ ìžˆë‹¤.

- `score_val` : ê²€ì¦ ë°ì´í„° ê¸°ì¤€ ì ìˆ˜ (í•™ìŠµ ì¤‘ ë‚´ë¶€ ê²€ì¦)
- `score_test` : ìš°ë¦¬ê°€ ë„£ì–´ì¤€ `test_df` ê¸°ì¤€ ì ìˆ˜  
- `pred_time_*`, `fit_time` : ì˜ˆì¸¡/í•™ìŠµ ì‹œê°„


````python
leaderboard = predictor.leaderboard(test_df, silent=False)
leaderboard
````

## 7. ì˜ˆì¸¡ & ì„±ëŠ¥ í‰ê°€

í•™ìŠµëœ AutoGluon ëª¨ë¸ë¡œ í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¥¼ ì˜ˆì¸¡í•´ ë³´ê³ ,  
ë‹¤ìŒê³¼ ê°™ì€ ì§€í‘œë¡œ ì„±ëŠ¥ì„ í‰ê°€í•œë‹¤.

- **ë¶„ë¥˜ ë¦¬í¬íŠ¸(classification_report)** : Precision, Recall, F1 ë“±
- **í˜¼ë™í–‰ë ¬(confusion_matrix)** : ì‹¤ì œ/ì˜ˆì¸¡ ì¡°í•©í‘œ
- **ROC-AUC ì ìˆ˜** : ì–‘ì„± í´ëž˜ìŠ¤(ì—°ë´‰ >50K) êµ¬ë¶„ ëŠ¥ë ¥


````python
# ì‹¤ì œ íƒ€ê¹ƒ
y_true = test_df['class']

# ì˜ˆì¸¡(0/1)
y_pred = predictor.predict(test_df)

# ì–‘ì„± í´ëž˜ìŠ¤(1)ì— ëŒ€í•œ ì˜ˆì¸¡ í™•ë¥ 
y_proba = predictor.predict_proba(test_df)[1]

print("===== Classification Report =====")
print(classification_report(y_true, y_pred))

print("\n===== Confusion Matrix =====")
print(confusion_matrix(y_true, y_pred))

print("\n===== ROC-AUC =====")
print(roc_auc_score(y_true, y_proba))
````

### 7.1 Confusion Matrix ì‹œê°í™”

í˜¼ë™í–‰ë ¬ì„ **ížˆíŠ¸ë§µ í˜•íƒœ**ë¡œ ì‹œê°í™”í•˜ë©´,  
ëª¨ë¸ì´ ì–´ëŠ í´ëž˜ìŠ¤ë¥¼ ë” ë§Žì´ í‹€ë¦¬ëŠ”ì§€ ì§ê´€ì ìœ¼ë¡œ ë³¼ ìˆ˜ ìžˆë‹¤.

- (0,0): ì‹¤ì œ 0ì„ 0ìœ¼ë¡œ ë§žì¶˜ ê°œìˆ˜ (TN)
- (0,1): ì‹¤ì œ 0ì„ 1ë¡œ ìž˜ëª» ì˜ˆì¸¡í•œ ê°œìˆ˜ (FP)
- (1,0): ì‹¤ì œ 1ì„ 0ìœ¼ë¡œ ìž˜ëª» ì˜ˆì¸¡í•œ ê°œìˆ˜ (FN)
- (1,1): ì‹¤ì œ 1ì„ 1ë¡œ ë§žì¶˜ ê°œìˆ˜ (TP)


````python
from matplotlib.colors import LogNorm

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(4,4))
plt.imshow(cm, interpolation='nearest')
plt.title("Confusion Matrix")
plt.colorbar()
tick_marks = np.arange(2)
plt.xticks(tick_marks, ['â‰¤50K', '>50K'])
plt.yticks(tick_marks, ['â‰¤50K', '>50K'])

plt.ylabel('ì‹¤ì œ ê°’')
plt.xlabel('ì˜ˆì¸¡ ê°’')

# ê° ì¹¸ì— ìˆ«ìž í‘œì‹œ
thresh = cm.max() / 2.
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, format(cm[i, j], 'd'),
                 ha="center", va="center",
                 color="white" if cm[i, j] > thresh else "black")

plt.tight_layout()
plt.show()
````

## 8. Feature Importance(ë³€ìˆ˜ ì¤‘ìš”ë„) ë¶„ì„

`predictor.feature_importance(test_df)`ë¥¼ ì‚¬ìš©í•˜ë©´  
**ì–´ë–¤ íŠ¹ì„±(feature)ì´ ì—°ë´‰ ì˜ˆì¸¡ì— ë” í° ì˜í–¥ì„ ë¯¸ì¹˜ëŠ”ì§€** ì•Œ ìˆ˜ ìžˆë‹¤.

ì´ ì •ë³´ë¥¼ í™œìš©í•˜ë©´:
- ì–´ë–¤ ë³€ìˆ˜ë“¤ì´ ì¤‘ìš”í•œì§€ ì„¤ëª…í•  ìˆ˜ ìžˆê³ 
- ë¶ˆí•„ìš”í•œ ë³€ìˆ˜ë¥¼ ì¤„ì´ê±°ë‚˜,
- ì¤‘ìš”í•œ ë³€ìˆ˜ì— ëŒ€í•´ ì¶”ê°€ ë¶„ì„(EDA)ì„ í•  ìˆ˜ ìžˆë‹¤.


````python
fi = predictor.feature_importance(test_df)
fi.head()
````

````python
# ì¤‘ìš”ë„ ê¸°ì¤€ìœ¼ë¡œ ì •ë ¬ í›„ ì‹œê°í™”
fi_sorted = fi.sort_values("importance", ascending=True)

plt.figure(figsize=(6, max(4, len(fi_sorted) * 0.3)))
plt.barh(fi_sorted.index, fi_sorted["importance"])
plt.title("AutoGluon Feature Importance (class: >50K)")
plt.xlabel("Importance")
plt.tight_layout()
plt.show()
````
