```python
def exrate_get(ymd_dash, ymd):
    url="https://www.kebhana.com/cms/rate/wpfxd651_01i_01.do"
    payload = dict(ajax="true",tmpInqStrDt=ymd_dash, pbldDvCd="0", inqStrDt=ymd, inqKindCd="1", requestTarget="searchContentDiv")
    r=requests.get(url,params=payload)
    df=pd.read_html(r.text)[0]
    df.insert(0, "날짜", ymd_dash)
    return df
```

# 멀티인덱스 컬럼명 단순화하기
* 0 != 1 == 2
* 0 != 1 != 2
* 0 == 1 == 2

```python
def flatten_cols(df):
    new_cols = {}
    for col in df.columns:
    #     print(col)
        if col[0] != col[1] == col[2]:
            if col[1]=="":
                new_cols[col]=col[0]
            else:
                new_cols[col]=col[0]+ "_" + col[1].replace(" ", "_")
        elif col[0] != col[1] != col[2]:
            new_cols[col]=" ".join(col).replace(" ", "_")
        else:
            new_cols[col]= col[0].replace(" ", "_")


    return new_cols
```

# 날짜 데이터 생성 후 데이터 수집하고 column명 변환해서 DB에 저장하기

```python
from io import StringIO
import requests
import pandas as pd
import time 
from datetime import datetime
from dbio import to_db
```

```python
# 하나은행 환율 수집 함수
def exrate_get(ymd_dash, ymd):
    url="https://www.kebhana.com/cms/rate/wpfxd651_01i_01.do"
    payload = dict(ajax="true",tmpInqStrDt=ymd_dash, pbldDvCd="0", inqStrDt=ymd, inqKindCd="1", requestTarget="searchContentDiv")
    r=requests.get(url,params=payload)
    df=pd.read_html(StringIO(r.text))[0]
    df.insert(0, "날짜", ymd_dash)
    return df
```

```python
# 컬럼명 평탄화하는 함수
def flatten_cols(df):
    new_cols = {}
    for col in df.columns:
    #     print(col)
        if col[0] != col[1] == col[2]:
            if col[1]=="":
                new_cols[col]=col[0]
            else:
                new_cols[col]=col[0]+ "_" + col[1].replace(" ", "_")
        elif col[0] != col[1] != col[2]:
            new_cols[col]=" ".join(col).replace(" ", "_")
        else:
            new_cols[col]= col[0].replace(" ", "_")


    return new_cols
```

```python
# 날짜 생성하고 데이터 수집하기
result=[]
for date in list(pd.date_range("2026-01-28", "2026-01-23",freq='-1B')):
    ymd_dash, ymd=date.strftime("%Y-%m-%d"), date.strftime("%Y%m%d")
    print(ymd_dash, ymd)
    data=exrate_get(ymd_dash, ymd)
    result.append(data)
    
final_result = pd.concat(result)

# 수집된 데이터프레임 컬럼명 평탄화하기
new_cols_dict=flatten_cols(final_result)
new_cols=[new_cols_dict[col] for col in final_result]
final_result.columns = new_cols
final_result

# dbio 함수로 결과 DB에 저장하기
to_db("exchange_rate_data", "exchange_rate", final_result)
```
