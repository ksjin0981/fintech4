
# 🧪 Ubuntu 24.04 실습 문제 모음 (SSH/서비스/방화벽 포함)

Ubuntu 기본 명령어, 권한, 서비스(SSH), 방화벽(UFW)를 연습하기 위한 **실습용 문제 모음**입니다.  
터미널에서 직접 명령어를 입력하며 풀어보세요.



## 1️⃣ 기초 명령어 실습 (파일/폴더, 조회, 이동)

### ✅ 실습 문제 1  
현재 작업 중인 디렉터리의 전체 경로를 출력하시오.

---

### ✅ 실습 문제 2  
현재 디렉터리에 존재하는 **모든 파일(숨김 파일 포함)** 을 **자세히(long format)** 출력하시오.

---

### ✅ 실습 문제 3  
`practice`라는 이름의 디렉터리를 생성하고, 그 안으로 이동하시오.

---

### ✅ 실습 문제 4  
현재 디렉터리 안에 `notes.txt`라는 빈 파일을 만들고, 그 파일을 `practice` 디렉터리 안으로 이동시키시오.

---

### ✅ 실습 문제 5  
현재 디렉터리에 있는 파일 목록을 **숨김 파일 포함, 자세히, 용량 단위를 보기 좋게** 출력하시오.

## 3️⃣ 권한/소유권 실습
