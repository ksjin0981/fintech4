```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import koreanize_matplotlib
```

```python
train_df = pd.read_csv("./data/ratings_train.txt", sep="\t")
train_df.head()
```

```python
test_df = pd.read_csv("./data/ratings_test.txt", sep="\t")
test_df.head()
```

정규표현식을 이용한 텍스트 정리

```python
import re # 정규표현식: 문자열을 원하는 규칙대로 자르고 찾고 검사하는 도구
def text_clean(text):
    cleaned = re.sub(r'[^가-힣a-zA-Z0-9]', " ", text) # 한글, 영어, 숫자를 제외한 모든 문자들을 공백으로 바꿔라
    cleaned = cleaned.replace("  ", " ").replace("  ", " ").strip() 
    return cleaned
```

```python
# apply: 함수를 받아서 각 행에 자동으로 실행시켜줌 SO 괄호 없이 함수 이름만 넘기면 됨
train_df['document'] = train_df['document'].apply(text_clean)
train_df.head()
```

```python
test_df['document'] = test_df['document'].apply(text_clean)
test_df.head()
```

* document에 있는 문자를 숫자로 변환
* 문자를 벡터화
* CountVectorizer()

```python
X = train_df['document']
y =  train_df['label']
```

```python
X
```

```python
y
```

```python
y.value_counts(normalize=True)   # stratify 옵션 안줘도 됨
```

```python
from sklearn.model_selection import train_test_split
```

```python
# test_size=0.7: 여기서는 valid 데이터를 70%로 쓰겠다는 뜻
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.7, random_state=42)
```

* 독립변수 X_train, X_valid를 countervectorizer로 숫자 변환
    * countervectorizer: 문장을 '숫자 벡터'로 바꿔주는 도구 (컴퓨터는 글자 이해 못하니까)
* 독립변수가 문자로 되어 있는 데이터의 경우 MultinomialNB

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB # 단어 등장 확률을 이용해서 문장을 분류하는 모델
from sklearn.metrics import classification_report
```

```python
pipe = Pipeline(steps=[
    ('cvec', CountVectorizer()),
    ('model', MultinomialNB())
])

pipe.fit(X_train, y_train)
pred = pipe.predict(X_valid)
print(classification_report(y_valid, pred))
```

* test 데이터 불러와서 검증

```python
X_test = test_df['document']
y_test = test_df['label']
```

```python
X_test
```

```python
y_test.value_counts(normalize=True)
```

```python
test_pred = pipe.predict(X_test)
print(classification_report(y_test, test_pred))
```

# CountVectorizer, TfidfVectorizer
### 1) CountVectorizer: 단어 빈도(횟수) 기반 벡터화
작동 방식: CountVectorizer는 각 문서에서 등장한 단어의 **횟수(빈도)**를 기준으로 벡터를 만듭니다. 단순히 각 문서에서 특정 단어가 몇 번 등장했는지를 세어 벡터화합니다.<br>
특징:
**각 단어의 빈도가 높을수록 해당 단어의 중요성이 더 크다고 가정**합니다.<br>
빈도가 높다는 것만을 고려하기 때문에, 모든 문서에서 자주 등장하는 단어도 중요한 단어로 처리될 수 있습니다.<br>
예시:<br>
예를 들어, 두 개의 문서가 있다면:<br>
문서 1: "고양이가 나무 위에 있다."<br>
문서 2: "나무 아래에 고양이가 있다."<br>
이 두 문서를 CountVectorizer로 변환하면 단어 빈도가 포함된 벡터가 생성됩니다:<br>
['고양이': 2, '나무': 2, '위에': 1, '아래에': 1, '있다': 2]<br>

```python
doc = ['고양이가 나무 위에 있다.', '나무 아래에 고양이가 있다.', '고양이 강아지가 나무 옆에 있다.']
c_vec = CountVectorizer()
t2vec = c_vec.fit_transform(doc)
print(c_vec.get_feature_names_out()) # 벡터화돤 단어 목록
print(t2vec.toarray()) # 단어 빈도 벡터
```

### 2) TfidfVectorizer: TF-IDF (Term Frequency-Inverse Document Frequency) 기반 벡터화
작동 방식: TfidfVectorizer는 단어의 빈도뿐만 아니라, 단어의 중요도를 계산합니다. 여기서는 TF-IDF 값을 사용하여 문서 간 차별성을 강조합니다.<br>
**TF (Term Frequency): 단어가 문서에서 얼마나 자주 등장했는지**를 나타냅니다.<br>
**IDF (Inverse Document Frequency): 단어가 다른 문서에 얼마나 자주 등장하지 않았는지**를 나타냅니다. 자주 등장하지 않는 단어는 더 중요한 단어로 간주합니다.<br>
특징:<br>
TF-IDF는 문서 전체에서 자주 등장하는 흔한 단어들(예: "그리고", "이다" 등)의 중요도를 낮추고, 문서에서만 중요한 단어들의 중요도를 높입니다.<br>
**단순히 빈도가 높은 단어보다 특정 문서에서 더 특징적인 단어에 더 높은 가중치를 부여**합니다.<br>
예시:<br>
위의 문서 1과 문서 2에 대해 TfidfVectorizer로 변환하면, 공통 단어들(예: "있다", "고양이")의 중요도는 낮아지고, 차별적인 단어(예: "위에", "아래에")의 중요도는 상대적으로 높아집니다.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
doc = ['고양이가 나무 위에 있다.', '나무 아래에 고양이가 있다.', '고양이 강아지가 나무 옆에 있다.']
tfidf_vec = TfidfVectorizer()
t2tfidf_vec = tfidf_vec.fit_transform(doc)
print(tfidf_vec.get_feature_names_out()) # 벡터화된 단어 목록
print(t2tfidf_vec.toarray()) # 단어 빈도 벡터
# TF * IDF 로 계산
```

### CountVectorizer와 TfidfVectorizer 비교

| **특성**             | **CountVectorizer**                                     | **TfidfVectorizer**                                               |
|----------------------|--------------------------------------------------------|-------------------------------------------------------------------|
| **기반**             | 단어의 단순 빈도                                        | 단어 빈도 * 문서 내에서의 상대적 중요도(TF-IDF)                     |
| **단어 빈도 계산**    | 문서에서 등장한 단어의 단순한 등장 횟수를 셈            | 단어의 등장 횟수(TF)와 해당 단어가 문서들에서 얼마나 자주 등장하지 않았는지를 함께 고려(IDF) |
| **빈번한 단어 처리**  | 문서에서 자주 등장하는 단어일수록 높은 가중치를 부여    | 문서에서 흔한 단어는 가중치를 낮추고, 드문 단어는 높은 가중치를 부여   |
| **주요 용도**        | 단순한 단어 빈도 기반 분석이 필요할 때 사용             | 문서 간 차별적인 단어를 구별할 때 유용                                |
| **계산 비용**         | 상대적으로 적음                                         | 상대적으로 더 복잡하고 계산 비용이 높음                                |


* tokenizer: 형태소를 분석해서 단어를 나누어 주는 역할, 기본은 공백을 기준으로 나눔
  * 한국어는 조사가 있기 때문에 공백이 아닌 형태소 분석을 통해서 나눠야 함.
  * konlpy의 형태소 분석기를 이용해서 형태소를 나누고 분석
* ngram_range(1, 2)
  * 단어를 벡터화 할 때 단어의 범위를 지정 (1, 2)는 1-gram, 2-gram
  * "이 영화는 정말 좋다"
  * 1-gram: ['이', '영화는', '정말', '좋다']
  * 2-gram: ['이 영화는', '영화는 정말', '정말 좋다']
* min_df=4 
  * 단어가 등장하는 최소 문서 수를 설정하는 파라미터
* max_df=0.9
  * 단어가 전체 문서의 90% 이하에서 등장할 때만 벡터화에 포함

* tokenizer: 문장을 어떻게 쪼갤지 정하는 함수 (기본은 공백을 기준으로 나눔)
* ngram_range(1,2): 한 단어만 볼지 (1-gram) 두 단어 묶음도 볼지 (2-gram)
* min_df=4: 최소 4개 문서 (한 줄) 이상에서 등장한 단어만 사용
* max_df=0.9: 전체 문서의 90% 이하에서 등장하는 단어만 포함

```python
# tokenizer를 konlpy의 mecab 사용, ngram=(1,2), min_df=4, max_df=0.9
pipe2 = Pipeline(steps=[
    ('cvec', CountVectorizer(tokenizer=mecab_tokenizer, token_pattern=None, 
                             ngram_range=(1,2), min_df=4, max_df=0.9)),
    ('model', MultinomialNB())
])

pipe2.fit(X_train, y_train)
pred2 = pipe2.predict(X_valid)
print(classification_report(y_valid, pred2))
# token_pattern=None: “기본 단어 추출 정규식을 비활성화하겠다”는 의미
```

```python
pred2 = pipe2.predict(X_test)
print(classification_report(y_test, pred2))
```

```python
# tokenizer를 TfidfVectorizer, konlpy의 mecab 사용, ngram=(1,2), min_df=4, max_df=0.9
pipe3 = Pipeline(steps=[
    ('cvec', TfidfVectorizer(tokenizer=mecab_tokenizer, token_pattern=None, 
                             ngram_range=(1,2), min_df=4, max_df=0.9)),
    ('model', MultinomialNB())
])

pipe3.fit(X_train, y_train)
pred3 = pipe3.predict(X_valid)
print(classification_report(y_valid, pred3))
```

```python
pred3 = pipe3.predict(X_test)
print(classification_report(y_test, pred3))
```

# random_forest와 비교

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# tokenizer를 TfidfVectorizer, konlpy의 mecab 사용, ngram=(1,2), min_df=4, max_df=0.9
pipe4 = Pipeline(steps=[
    ('cvec', TfidfVectorizer(tokenizer=mecab_tokenizer, token_pattern=None, 
                             ngram_range=(1,2), min_df=4, max_df=0.9)),
    ('model', RandomForestClassifier(random_state=42, n_jobs=-1))
])

pipe4.fit(X_train, y_train)
pred4 = pipe4.predict(X_valid)
print(classification_report(y_valid, pred4))
```

```python
pred4 = pipe4.predict(X_test)
print(classification_report(y_test, pred4))
```
